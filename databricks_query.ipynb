```sql
%sql
-- S3'ten veriyi okuyup 'amazon_gold_sales' adında kalıcı bir tablo oluşturur.
-- Sütun isimlerindeki boşluk hatalarını (Column Mapping) bu komutla engelliyoruz.
CREATE TABLE IF NOT EXISTS amazon_gold_sales 
TBLPROPERTIES ('delta.columnMapping.mode' = 'name') 
AS 
SELECT * FROM read_files(
  's3://faruk-amazon-data/Amazon Sale Report.csv',
  format => 'csv',
  header => true,
  inferSchema => true
);
```

```sql
%sql
SELECT * FROM amazon_tablosu LIMIT 10;
```

```sql
%sql
-- En çok sipariş alan (satış adedi en yüksek) kategoriyi bulur.
SELECT Category, COUNT(*) AS total_orders
FROM amazon_gold_sales
GROUP BY Category
ORDER BY total_orders DESC
LIMIT 1;
```

```sql
%sql
-- Hangi eyaletin (state) toplamda daha fazla para (Amount) getirdiğini hesaplar.
SELECT `ship-state` AS state, ROUND(SUM(Amount), 2) AS total_revenue
FROM amazon_gold_sales
GROUP BY state
ORDER BY total_revenue DESC
LIMIT 1;
```

```sql
%sql
-- Toplam kazancı toplam ürün adedine bölerek sepet ortalamasını bulur.
SELECT ROUND(SUM(Amount) / SUM(Qty), 2) AS average_basket_value
FROM amazon_gold_sales;
```

```sql
%sql
-- This query calculates the top 10 cities by total revenue.
-- ROUND function is used to keep the currency format clean (2 decimal places).
SELECT 
    `ship-city` AS city, 
    ROUND(SUM(Amount), 2) AS total_sales
FROM amazon_tablosu
GROUP BY `ship-city`
ORDER BY SUM(Amount) DESC
LIMIT 10;
```
